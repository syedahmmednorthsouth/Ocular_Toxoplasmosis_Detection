```python
import os
import pandas as pd

# Path to the folder containing images and masks
folder_path = 'data'

# Create a list to store the data
data = []

# Iterate through the files in the images folder
for filename in os.listdir(os.path.join(folder_path, 'images')):
    if filename.endswith(".jpg"):  # Assuming your images are in jpg format
        image_path = os.path.join(folder_path, 'images', filename)

        # Extract the image number (assuming the image filename is a number)
        image_number = filename.split('.')[0]

        # Construct the mask filename with variations
        mask_filename = f"{image_number}.jpg"

        # Check if the mask file exists
        mask_path = os.path.join(folder_path, 'masks', mask_filename)
        if not os.path.exists(mask_path):
            # If the exact match doesn't exist, try with variations
            mask_filename_alternative = f"{image_number}-a.jpg"
            mask_path_alternative = os.path.join(folder_path, 'masks', mask_filename_alternative)

            if os.path.exists(mask_path_alternative):
                mask_path = mask_path_alternative

        # Check if the image file exists in the mask folder before appending
        if os.path.exists(mask_path):
            data.append([image_path, mask_path])

# Create a DataFrame from the list
df = pd.DataFrame(data, columns=['image_path', 'mask_path'])

# Display the DataFrame
print(df)
```

```python
import pandas as pd
import cv2

# Function to load image and mask
def load_image_mask(image_path, mask_path):
    image = cv2.imread(image_path)
    mask = cv2.imread(mask_path)

    if image is None or mask is None:
        print(f"Error loading image or mask: {image_path}, {mask_path}")
        return None, None

    # Convert BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)

    return image, mask

# Assuming raw_df is your DataFrame
for index, row in raw_df.iterrows():
    image, mask = load_image_mask(row['image_path'], row['mask_path'])

    if image is None or mask is None:
        # Remove the row from the DataFrame if an error occurs
        raw_df = raw_df.drop(index, axis=0)

# Optionally, you can reset the index of the DataFrame after removing rows
raw_df = raw_df.reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

def split_dataframe_by_paths(df, test_size=0.2, random_state=36):
    # Split df into train and val
    train_df, val_df = train_test_split(df,
                                        test_size=test_size,
                                        random_state=random_state)

    # Reset indices
    train_df, val_df = train_df.reset_index(drop=True), val_df.reset_index(drop=True)

    return train_df, val_df

# Usage
train_df, val_df = split_dataframe_by_paths(df)

print("Train df: {}\nVal df: {}\n"\
      .format(train_df.shape, val_df.shape))
```

```python
import torch
import torch.nn as nn

# Define the U-Net architecture
class UNet(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UNet, self).__init__()
        
        # Encoder (contracting path)
        self.encoder = nn.Sequential(
            nn.Conv2d(in_channels, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            # nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            # nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            # nn.MaxPool2d(kernel_size=2, stride=2),
        )
        
        # Bottleneck
        self.bottleneck = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
        )
        
        # Decoder (expansive path)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            
            nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            
            nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2),
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
        )
        
        # Output layer
        self.out = nn.Conv2d(64, out_channels, kernel_size=1)
    
    def forward(self, x):
        # Encoder
        enc1 = self.encoder(x)
        # Bottleneck
        bottleneck = self.bottleneck(enc1)
        # Decoder
        dec1 = self.decoder(bottleneck)
        # Output
        out = self.out(dec1)
        return out

# Define the number of input channels and output channels for your segmentation task
in_channels = 3  # For RGB images
out_channels = 1  # For binary segmentation

# Create an instance of the U-Net model
model = UNet(in_channels, out_channels)

# Print the model architecture
print(model)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import jaccard_score, f1_score
from torch.utils.data import DataLoader

# Move the model to the GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Define your U-Net model (as shown in the previous code)

# Define hyperparameters
learning_rate = 0.001
epochs = 10

# Define loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Create DataLoader instances for training and validation (as shown in the previous code)
# train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=True)
# val_dataloader = DataLoader(val_dataset, batch_size=4, shuffle=False)

# Training loop
for epoch in range(epochs):
    model.train()
    total_loss = 0.0
    for batch in train_dataloader:
        images, masks = batch['image'].to(device), batch['mask'].to(device)
        images = images.permute(3, 1, 2, 0)
        # From: [batch_size, depth, height, width, channels]
        # To: [batch_size, channels, depth, height, width]
        # input = input.permute(0, 4, 1, 2, 3)
        
        outputs = model(images)
        
        # Calculate the loss
        loss = criterion(outputs, masks)
        
        # Backpropagation and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    # Print training loss for the epoch
    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {total_loss / len(train_dataloader)}")

    # Validation
    model.eval()
    with torch.no_grad():
        # Define empty lists to store prediction masks and target masks
        all_preds = []
        all_targets = []
        
        for batch in val_dataloader:
            images, masks = batch['image'].to(device), batch['mask'].to(device)
            
            # Forward pass
            outputs = model(images)
            
            # Apply a threshold to the model outputs to convert them to binary masks
            preds = (outputs > 0.5).float()
            
            # Convert PyTorch tensors to NumPy arrays for evaluation
            preds_np = preds.cpu().numpy()
            targets_np = masks.cpu().numpy()
            
            all_preds.append(preds_np)
            all_targets.append(targets_np)
        
        # Calculate IoU and Dice coefficient metrics
        all_preds = np.concatenate(all_preds)
        all_targets = np.concatenate(all_targets)
        
        iou = jaccard_score(all_targets.flatten(), all_preds.flatten())
        dice = f1_score(all_targets.flatten(), all_preds.flatten())
        
        print(f"Validation - IoU: {iou}, Dice: {dice}")

# Save the trained model
torch.save(model.state_dict(), 'unet_segmentation_model.pth')
```
